```python
import requests
import json

AGENT_URL = "http://aython-agent:4000"
headers = {"Content-Type": "application/json"}

# Initialize agent
payload = {
    "jsonrpc": "2.0",
    "method": "init_agent",
    "params": {"model": "gpt-4o-mini"},
    "id": 1
}
response = requests.post(AGENT_URL, headers=headers, data=json.dumps(payload))
print(response.json())

# Generate and run code
payload = {
    "jsonrpc": "2.0",
    "method": "generate_and_run",
    "params": {"requirements": "create a function to add two numbers"},
    "id": 2
}
response = requests.post(AGENT_URL, headers=headers, data=json.dumps(payload))
print(response.json())
```
